```python
class Category:
  BOOKS = "BOOKS"
  CLOTHING = "CLOTHING"

train_x =["i love the book", "this is a great book", "this fit is great", "i love the shoes"]
train_y = [Category.BOOKS, Category.BOOKS, Category.CLOTHING, Category.CLOTHING]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(binary=True) # vectorizer = CountVectorizer(binary=True, ngram_range=(1,2)) -> kelimeleri ikişer ikişer ele alacak range ifadesi ile
train_x_vectors = vectorizer.fit_transform(train_x)

print(train_x_vectors[0])
```

```python
vectorizer.get_feature_names_out()
```

```python
train_x_vectors.toarray()
```

```python
from sklearn import svm

clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(train_x_vectors, train_y)
```

```python
test_x= vectorizer.transform(['i love the book'])

clf_svm.predict(test_x)
```

```python
test_x2= vectorizer.transform(['i love the shoes'])

clf_svm.predict(test_x2)
```

```python
test_x3= vectorizer.transform(['i love the novel'])

clf_svm.predict(test_x3) # Bilmediği kelimelerde çok kötü sonuç veriyor.
```

##**REGEX**

```python
import re
regexp = re.compile(r"ab[^\s]*cd$")
phrases = ["abcd", "xxx", "abxxxcd", "ab cd"]
matches = []
for phrase in phrases:
  if re.match(regexp, phrase):
    matches.append(phrase)

print(matches)
```

```python
import re
regexp = re.compile(r"read|story|book")
phrases = ["I liked that history", "the car threaded up the hill", "this hat is nice"]
matches = []
for phrase in phrases:
  if re.search(regexp, phrase):
    matches.append(phrase)

print(matches)
```

##**Stemming/Lemmatization**

```python
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Stemming
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

phrase = "reading the books"
words = word_tokenize(phrase)

stemmed_words = []
for word in words:
  stemmed_words.append(stemmer.stem(word))

" ".join(stemmed_words)
```

```python
# Stemming
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

phrase = "stories"
words = word_tokenize(phrase)

stemmed_words = []
for word in words:
  stemmed_words.append(stemmer.stem(word))

" ".join(stemmed_words)  # Stori diyerek yanlış bir çözüm elde ederiz bu yöntem garanti bir yöntem değildir
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

phrase = "reading the books"
words = word_tokenize(phrase)

lemmatized_words = []

for word in words:
  lemmatized_words.append(lemmatizer.lemmatize(word))

" ".join(lemmatized_words)
```

##**Stopword Removal**

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)
```

```python
len(stop_words)
```

```python
phrase = "Here is an example sentence demonstrating the removal of stopwords"
words = word_tokenize(phrase)

stripped_phrase = []
for word in words:
  if word not in stop_words:
    stripped_phrase.append(word)

" ".join(stripped_phrase)
```

```python
import spacy, torch
nlp = spacy.load('en_core_web_md')
doc = nlp("Here is some text to encode.")
```

```python
class Category:
  BOOKS = "BOOKS"
  BANK = "BANK"

train_x =["good characters and plot progression", "check out the book", "good story. would recommend","novel recommandation" , "need to make a deposit to the bank", "balance inquiry savings", "save money"]
train_y = [Category.BOOKS, Category.BOOKS, Category.BOOKS, Category.BOOKS, Category.BANK, Category.BANK, Category.BANK]
```

```python
from sklearn import svm

docs = [nlp(text) for text in train_x]
train_x_vectors = [doc.vector for doc in docs]

clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(train_x_vectors, train_y)

test_x = ["i need to write a check"]
docs = [nlp(text) for text in test_x]
test_x_vectors = [doc.vector for doc in docs]

clf_svm.predict(test_x_vectors) # RNN ile doğru sonuç verme ihtimali çok artar
```
