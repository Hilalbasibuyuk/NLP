```python
class Category:
  BOOKS = "BOOKS"
  CLOTHING = "CLOTHING"

train_x =["i love the book", "this is a great book", "this fit is great", "i love the shoes"]
train_y = [Category.BOOKS, Category.BOOKS, Category.CLOTHING, Category.CLOTHING]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(binary=True) # vectorizer = CountVectorizer(binary=True, ngram_range=(1,2)) -> kelimeleri ikişer ikişer ele alacak range ifadesi ile
train_x_vectors = vectorizer.fit_transform(train_x)

print(train_x_vectors[0])
```

```python
vectorizer.get_feature_names_out()
```

```python
train_x_vectors.toarray()
```

```python
from sklearn import svm

clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(train_x_vectors, train_y)
```

```python
test_x= vectorizer.transform(['i love the book'])

clf_svm.predict(test_x)
```

```python
test_x2= vectorizer.transform(['i love the shoes'])

clf_svm.predict(test_x2)
```

```python
test_x3= vectorizer.transform(['i love the novel'])

clf_svm.predict(test_x3) # Bilmediği kelimelerde çok kötü sonuç veriyor.
```

```python
import spacy
nlp = spacy.load('en_core_web_md')
```

```python
print(train_x)
```

```python
docs = [nlp(text) for text in train_x]
print(docs)
```

```python
print(docs[0].vector)
```

```python
train_x_word_vectors = [x.vector for x in docs]
```

```python
clf_svm_wv = svm.SVC(kernel='linear')
clf_svm_wv.fit(train_x_word_vectors, train_y)
```

```python
test_x = ["i love the story"]
test_docs = [nlp(text) for text in test_x]
test_x_word_vectors = [x.vector for x in test_docs]

clf_svm_wv.predict(test_x_word_vectors)
```

```python
test_x = ["i love the novel"]
test_docs = [nlp(text) for text in test_x]
test_x_word_vectors = [x.vector for x in test_docs]

clf_svm_wv.predict(test_x_word_vectors)
```

```python
test_x = ["i love the hat"]
test_docs = [nlp(text) for text in test_x]
test_x_word_vectors = [x.vector for x in test_docs]

clf_svm_wv.predict(test_x_word_vectors) #word vector sayesinde benzer kelimeleri gruplandırabiliyor.
```
